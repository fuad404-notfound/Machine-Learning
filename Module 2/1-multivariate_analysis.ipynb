```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_data_updated.csv')
df.head()
```

# Multivariate Analysis- 1 ( Countplot with Hue )

```python
sns.countplot(x=df['Sex'],hue=df['Survived'])
```

```python
gender_split = df.groupby('Sex')['Survived'].value_counts()

print(gender_split)
```

```python
# for percentange
gender_split = df.groupby('Sex')['Survived'].value_counts(normalize=True) * 100

print(gender_split)
```

# Multivariate Analysis -2 ( Countplot with Hue-2)

```python
sns.countplot(x=df['Pclass'],hue=df['Survived'])
```

```python
pclass_split = df.groupby('Pclass')['Survived'].value_counts(normalize=True) * 100

print(pclass_split)
```
