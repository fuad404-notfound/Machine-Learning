```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_data_updated.csv')
df.head(10)
```

# Univariate Analysis( Categorical Value Analysis -1 )

```python
sns.countplot(data=df, x='Survived')
```

```python
df['Survived'].value_counts()
```

```python
(df['Survived'].value_counts() / len(df)) * 100
```

```python
survived_count = df['Survived'].value_counts()

survived_label = df['Survived'].unique()

# print(survived_label)

plt.pie(survived_count, labels=survived_label.tolist(), autopct="%.2f")

plt.title('Pie Chart of Survival Rate')

plt.show()
```

# Univariate Analysis ( Numerical Column Analysis - 1)

```python
sns.histplot(np.array(df['Age']),bins=50)

plt.title('Histogram of Age')
plt.xlabel('Ages')
plt.ylabel('Frequency')

plt.show()
```

```python
sns.kdeplot(np.array(df['Age']))
plt.title('KDE plot of Age')
plt.xlabel('Ages')
plt.ylabel('Probabilty distribution')

plt.show()
```
